```python
import sys        
 
# appending the directory of mod.py
# in the sys.path list
sys.path.append('D:\CESI\A3\Bloc Algorithmique avancee\Projet\Git\projet_algo') 

import dataset as dataset
from colin.stats_Colin import histogramme,courbe,courbe_stats
import utils.prog_lineraire as prog_lineaire
import algorithmes.aco as aco
import statistics
import numpy as np

debut_ant = 1
fin_ant = 10

bornes = []


constraint = {
    "start":4,
    "city":5,
    "inferior_dist":1000,
    "superior_dist":2000,
    "enabled" : True,
}
moyennes = []
deviations = []

for taille_fourmi in range(debut_ant, fin_ant):
    parameters = {
    "nbr_ant":taille_fourmi,    
    "a":1.75,
    "b":3.75,
    "p":0.1,
    "nb_iter_max":20,
    "time_max":5,
    "nbr_random_search":10,
    "nbr_loops":5,
}
    for i in range(5):
        graphe_complet, cities, cities_coordonates = dataset.import_dataset("../dataset/dataset"+str(i))
        borne = prog_lineaire.prog_lineaire(graphe_complet, cities)[-1]
        res = []
        aco.aco_algorithm(graphe_complet,parameters,res,constraint)
        best_path_value = res[0][2]
        instance = borne*100/best_path_value
        bornes.append(instance)

    moyennes.append(statistics.fmean(bornes))               
    deviations.append(np.std(bornes))
```
